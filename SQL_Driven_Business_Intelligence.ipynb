```python
!pip install pandasql
import pandas as pd
import pandasql as ps

# 1. Load the data
df = pd.read_csv('Electonic Shop Dataset.csv')

# 2. SQL Query: Finding high-value customers
# Hum un customers ko dhoond rahe hain jinhone $1000 se zyada ki shopping ki hai
query = """
SELECT customer_name, location, product_name, (product_price * number_of_ordered_items) as total_spent
FROM df
WHERE (product_price * number_of_ordered_items) > 1000
ORDER BY total_spent DESC
LIMIT 10
"""

# Run the query
high_value_customers = ps.sqldf(query, locals())
print("--- TOP 10 HIGH VALUE TRANSACTIONS ---")
print(high_value_customers)
```

```python
query_2 = """
SELECT product_name, COUNT(*) as cancel_count
FROM df
WHERE order_status = 'Canceled'
GROUP BY product_name
ORDER BY cancel_count DESC
LIMIT 5
"""

canceled_products = ps.sqldf(query_2, locals())
print("--- TOP 5 CANCELED PRODUCTS ---")
print(canceled_products)
```

```python
import pandas as pd
import pandasql as ps

# 1. Load the data
df = pd.read_csv('Electonic Shop Dataset.csv')

# --- QUERY 1: Total Revenue per Product ---
query_revenue = """
SELECT product_name, SUM(product_price * number_of_ordered_items) as total_revenue
FROM df
GROUP BY product_name
ORDER BY total_revenue DESC
LIMIT 5
"""

# --- QUERY 2: Most Canceled Products (Problem Area) ---
query_canceled = """
SELECT product_name, COUNT(*) as cancel_count
FROM df
WHERE order_status = 'Canceled'
GROUP BY product_name
ORDER BY cancel_count DESC
LIMIT 5
"""

# Executing Queries
revenue_res = ps.sqldf(query_revenue, locals())
canceled_res = ps.sqldf(query_canceled, locals())

print("--- TOP 5 REVENUE GENERATING PRODUCTS ---")
print(revenue_res)
print("\n--- TOP 5 MOST CANCELED PRODUCTS ---")
print(canceled_res)
```
